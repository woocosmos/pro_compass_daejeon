# VIF(분산팽창요인)

```python
features = 'sg_walk_cnt+sg_car_cnt+safe_zn_cnt+cro_cnt+spd_sign_cnt+spd_diff+spd_sign+pull_over_cnt+cctv_cnt+pop+old_pro+kid_pro+building_kid_cnt+building_old_cnt'
```

```python
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_data = data
y, X = dmatrices('acci_cnt ~'+features, data=X_data, return_type='dataframe')

vif = pd.DataFrame()
vif['VIF Factor'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['features'] = X.columns

vif.round(1)
# intercept는 기본적으로 들어감
```

# 잔차 분석

## 전체 교통사고

```python
from statsmodels.formula.api import ols

formula = 'acci_cnt ~ sg_car_cnt + safe_zn_cnt + spd_sign_cnt + spd_diff + pull_over_cnt + cctv_cnt + pop'

res = ols(formula, data=data).fit()

fitted = res.predict(data)
residual = data['acci_cnt'] - fitted
sns.regplot(fitted, residual, lowess=True, line_kws={'color': 'red'})
plt.plot([fitted.min(), fitted.max()], [0, 0], '--', color='grey')
```

```python
import scipy.stats
sr = scipy.stats.zscore(residual)
(x, y), _ = scipy.stats.probplot(sr)
sns.scatterplot(x, y)
plt.plot([-3, 3], [-3, 3], '--', color='grey')
```

```python
scipy.stats.shapiro(residual)
```

```python
sns.regplot(fitted, np.sqrt(np.abs(sr)), lowess=True, line_kws={'color': 'red'})
```
